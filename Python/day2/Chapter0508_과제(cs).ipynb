[문제1]

Python의 변수 할당 개념을 다른 언어와 비교해 설명해주세요


x, y 같은 변수(Variables)에 이름 혹은 값을 지정해서 메모리 공간에 저장하는 개념입니다.

x는 이름을 지어주기 전까지는 이름이 없습니다. 이대로 출력하면 정의된 게 없다고 에러만 뜨는데, 여기에 10이라는 값을 혹은 "개발자"라는 이름을 정의해주면, x를 출력할 때 할당 즉, 저장한 값 혹은 이름이 나옵니다.

*None을 할당해서 빈 변수를 만들 수 있긴 하지만 아무것도 출력되지 않습니다.


파이썬과 가장 대조적인 컴파일 대표 C언어와 비교해보곘습니다.
파이썬은 객체 지향(Object-Oriented Programming, OOP) 언어로, 변수에 객체를 할당하여 그 객체값 혹은 이름을 참조합니다. 파이썬 내의 모든 것은 객체입니다.

C 언어는 대표적인 절차지향(Procedure-Oriented Programming, POP) 기법을 활용합니다.컴퓨터의 처리구조와 유사해 실행 속도가 빠른 장점이 있습니다.

1. 정적 타이핑 vs 동적 타이핑
C언어는 정적 타이핑 언어로, 변수를 선언할 때 명시적으로 타입을 지정해야 합니다.(=변수를 만들 때 미리 어느 데이터를 담을지 명시). 그에 반해, 파이썬은 동적 타이핑 언어로, 변수의 타입이 런타임에 동적으로 결정됩니다.(미리 정하지 않아도 나중에 어떤 데이터를 넣어도 괜찮습니다.).

2. 메모리 관리
C언어는 개발자가 메모리 할당, 해제를 수동으로(=직접) 관리해야 합니다. 즉, 직접 방을 치우고 정돈해야 한다는 의미 입니다.
파이썬은 동적으로 메모리를 할당하고 가비지 컬렉션(Garbage Collection)에 의해 자동으로 해제됩니다. 즉, 청소 자동화 봇이 있다는 의미 입니다.

3. 레퍼런스 및 포인터
C 언어는 포인터를 사용하여 메모리 주소를 직접 조작할 수 있습니다.변수가 직접 데이터 주소를 갖고 있어 값을 변경하면 그 주소값도 바뀌는 것과 같습니다.
파이썬은 변수는 값 대신에 객체에 대한 레퍼런스를 가지고 있습니다. 데이터의 주소값을 직접 알려주는 것보다 그 위치를 알려주는 것과 같습니다. 그래서 값을 바꾸면 주소값이 아닌 위치값이 변하는 거라고 볼 수 있습니다.

4. 여러 변수에 동시 할당
C 언어는 여러 변수에 동시에 값을 할당하기 위해서는 여러 줄의 코드가 필요합니다. 각 변수에 값을 따로 넣어줘야 해서 한 번에 많은 일을 하기 어렵습니다.
파이썬은 여러 변수에 동시에 값을 할당할 수 있어서 멀티 태스킹이 가능합니다. C언어가 일대일만 가능한 메신저라면, 파이썬은 단톡으로 한번에 전달 가능하다고 볼 수 있습니다.

[ 요약 ]
파이썬은 유연하고 편리한 문법을 제공하지만, C 언어는 더 직접적이고 저수준의 메모리 조작을 할 수 있습니다. C 언어는 정직하게 절차를 밟아가며 부지런하게 직접 일을 수행하는 반면 파이썬은 편리하게 많은 일을 자동으로 처리해서 프로그램으로 볼 수 있습니다.
