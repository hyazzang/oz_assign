[문제6]

ZeroDivisionError,  IndexError 외에 3가지 이상의 파이썬 에러에 대해 설명해주세요

```python
# TypeError
- 지원되지 않는 연산을 수행하거나, 잘못된 데이터 타입의 객체를 사용 할 때 발생
- 해결 방법: 타입을 맞춰 연산

# ValueError
- 올바른 데이터 타입이지만, 값이 부적절 할 때 발생
- 해결 방법: 입력값이 적절한지 확인, 필요한 경우 변환 전 확인/예외 처리

# KeyError
- 딕셔너리에서 존재하지 않는 키에 접근하려고 할 때 발생
- 해결 방법: 딕셔너리 키를 확인하고 키가 없는 경우 기본값을 반환하도록 설정

# AttributeError
- 객체가 호출하려는 속성이나 메소드를 가지고 있지 않을때 발생
- 해결 방법: 객체에 속성이나 메소드가 정의되어 있는지 확인

# NameError
- 정의되지 않은 변수나 함수에 접근하려고 할 때 발생
- 해결 방법: 변수를 정의한 후 사용하거나 오타를 확인
```
