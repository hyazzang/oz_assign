설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python
class BungeoppangShop: #붕어빵가게 클래스 정의
    def __init__(self):
        # 초기 재고
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }
        # 가격
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }
        # 판매량
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }

    def check_stock(self): #check_stock 이라는 함수 정의
        print("현재 재고:")
        for bread, quantity in self.stock.items(): # self.stock "키":값
            print(f'{bread}: {quantity}개') # 붕어빵 종류 : 붕어빵 재고

    def process_order(self, bread_type, bread_count): # process_order 함수 정의

        # 잘못된 붕어빵
        if bread_type not in self.stock: # self.stock 변수에 bread_type 없다면
            print(f'없는 붕어빵 입니다! {bread_type}')
            return

        # 재고 확인
        if self.stock[bread_type] < bread_count: # 붕어빵 종류의 재고가 주문 수량보다 미만
            print(f"재고 부족! {bread_type}의 재고는 {self.stock[bread_type]}개입니다.")
            return

        # 주문
        self.stock[bread_type] -= bread_count  # 붕어빵 종류의 재고 -
        self.sales[bread_type] += bread_count  # 붕어빵 종류의 판매량 +
```

함수로 리팩토링한 "붕어빵 자동 판매 시스템" 작성 후 실행합니다.

```python
class BungeoppangShop: # 붕어빵가게 클래스 정의
    def __init__(self):
        # 초기 재고
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }
        # 가격
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }
        # 판매량
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }

    def check_stock(self): # check_stock 함수 정의
        print("현재 재고:")
        for bread, quantity in self.stock.items(): # self.stock "키":값
            print(f'{bread}: {quantity}개') # 붕어빵 종류 : 붕어빵 재고

    def process_order(self, bread_type, bread_count): # process_order 함수 정의

        # 잘못된 붕어빵
        if bread_type not in self.stock: # self.stock 변수에 bread_type 없다면
            print(f'없는 붕어빵 입니다! {bread_type}')
            return

        # 재고 확인
        if self.stock[bread_type] < bread_count: # 붕어빵 종류의 재고가 주문 수량보다 미만
            print(f"재고 부족! {bread_type}의 재고는 {self.stock[bread_type]}개입니다.")
            return

        # 주문
        self.stock[bread_type] -= bread_count  # 붕어빵 종류의 재고 -
        self.sales[bread_type] += bread_count  # 붕어빵 종류의 판매량 +
        total_price = self.prices[bread_type] * bread_count # 총 가격 + 붕어빵종류 가격 * 주문 수량
        print(f'주문완료! {bread_type} {bread_count}개, {total_price}원 입니다.')

    def admin_mode(self, bread_type, additional_stock): # admin_mode 함수 정의

        # 잘못된 붕어빵
        if bread_type not in self.stock: # self.stock 변수에 bread_type 없다면
            print(f'없는 붕어빵 입니다! {bread_type}')
            return

        # 잘못된 재고
        if additional_stock <= 0: # 추가된 재고가 0 이하라면
            print("재고는 1개 이상 추가해야 합니다.")
            return

        # 재고 추가
        self.stock[bread_type] += additional_stock # 붕어빵 종류의 재고를 추가된 재고만큼 +

    def calculate_total_sales(self): # calculate_total_sales 함수 정의
        total_sales = 0  # 총 매출 0 값
        # 판매량 안의 빵 종류, 판매수량이 True 일때
        for bread, quantity_sold in self.sales.items():
            total_sales += self.prices[bread] * quantity_sold # 총 매출 + 빵종류 가격 + 판매수량
        print(f"총 매출: {total_sales}원")
        return total_sales # total_sales에 반환

def main(): # main 함수 정의
    shop = BungeoppangShop()
    print("붕어빵 자동 판매 시스템")

    while True:
        mode = input("모드 선택: (1.주문모드, 2.관리자모드, 3.종료)")

        if mode == "1": # 1 입력시 주문 모드
            bread_type = input("주문 붕어빵 종류 입력: ")
            bread_count = int(input("주문 수량 입력: "))
            shop.process_order(bread_type, bread_count)

        elif mode == "2": # 2 입력시 관리자 모드
            bread_type = input("추가할 붕어빵 종류 입력: ")
            additional_stock = int(input("추가할 재고 수량 입력: "))
            shop.admin_mode(bread_type, additional_stock)

        elif mode == "3": # 3 입력시 종료
            print("잘가용")
            shop.calculate_total_sales()
            break
        else:
            print("잘못된 입력입니다. 다시 입력해주세요.")

if __name__ == "__main__":
    main()
```
